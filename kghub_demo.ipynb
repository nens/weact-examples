# We-ACT KG-Hub Demo notebook
This notebook is used to demonstrate the use of the KG-hub, the datawarehouse deployed in Kyrgyzstan as part of the We-Act project. The datawarehouse can be used to store and retrieve timeseries data in Kyrgyzstan. This demo shows how Python can be used to get data from the KG-hub and plot in on a map.

```python
# Start by importing required packages
import requests
from get_functions import get_organisation_uuid, get_locations_list, get_timeseries_list, get_timeseries_events
from plot_functions import plot_points, plot_timeseries
```

```python
# Find the uuid

org_name = "Nelen & Schuurmans"

org_uuid = get_organisation_uuid(org_name)

print(f"Found organisation uuid {org_uuid}")
```

```python
# Show the available locations

locations = get_locations_list(org_uuid)
print(f"Found {len(locations)} locations")

plot_points(locations)

```

```python
# Show the available timeseries

timeseries = get_timeseries_list(org_uuid)
print(f"Found {len(timeseries)} timeseries")

timeseries_uuid = timeseries.loc[0,"uuid"]
```

```python
# Get and plot timeseries events

start = "2024-11-01T00:00:00"
end = "2024-11-06T00:00:00"

events = get_timeseries_events(timeseries_uuid, start, end)
plot_timeseries(events)
```
